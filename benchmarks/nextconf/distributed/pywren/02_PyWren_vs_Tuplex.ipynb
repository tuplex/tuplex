```python
df_tuplex = pd.read_csv('tuplex_results.csv')
df_tuplex['framework'] = 'Tuplex'
```

```python
df_pywren = pd.read_csv('pywren_results.csv')
df_pywren['framework'] = 'PyWren'
```

```python
df = pd.concat((df_tuplex, df_pywren))
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')

sns.factorplot(data=df,
            y='job_time', x='size',
            hue='framework', col='type', kind='bar', capsize=0.1)
plt.xlabel('size in MB')
plt.ylabel('end-to-end runtime in s')
```
