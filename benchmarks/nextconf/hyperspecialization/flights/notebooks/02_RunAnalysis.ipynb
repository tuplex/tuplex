```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import os
import glob
import json
import csv
import sys
import numpy as np
import math
import functools
from tqdm import tqdm
```

```python
paths = glob.glob('../build/benchmark_results/*.txt')
```

```python
rows = []

for path in paths:
    run = int(path[path.find('run-') + 4:path.rfind('.')])
    name = os.path.basename(path)
    mode = name[len('flights-'):name.rfind('-run')]
    
    content = open(path, 'r').readlines()
    
    row = {'name' : name, 'mode' : mode, 'run': run}
    row.update(json.loads(content[-1]))
    
    ns_to_s = 1000000000
    row['output'] /= ns_to_s
    row['transform'] /= ns_to_s
    row['total'] /= ns_to_s
    rows.append(row)
    
df = pd.DataFrame(rows)
df.head()
```
