```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# this is a mockup plot

global_data = {'normal' : 1025, 'general':100024, 'fallback': 73657 }


per_year_data = {2011 : {'normal' : 14000, 'general': 5000, 'fallback': 200},
                2012 : {'normal' : 12000, 'general': 700, 'fallback': 0},
                2013 : {'normal' : 9000, 'general': 2500, 'fallback': 50},}
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.bar(global_data.keys(), global_data.values())
plt.subplot(122)

dk = []
dv = []
for k in per_year_data.keys():
    d = per_year_data[k]
    dk += [str(k) + x for x in list(d.keys())]
    dv += list(d.values())
plt.bar(dk, dv)
```

```python
# -> per year brings more data onto the normal path & lowers overall memory footprint!
```
